```python
import pandas as pd
import numpy as np
```

```python
mortality_df = pd.read_csv("Resources/mortality_county.csv")
pollution_df = pd.read_csv("Resources/pollution_clean.csv")
```

```python
# clean mortality data years 2006 to 2010
mortality_2006_2010 = mortality_df[mortality_df["Year"] < 2011]
mortality_2006_2010
```

```python
mortality_county_ct = mortality_2006_2010["County"].unique().tolist()
print(mortality_county_ct, len(mortality_county_ct))
```

```python
mortality_2006_2010.to_csv("Resources/mortality_2006_2010.csv")
```

```python
pollution_2006_2010 = pollution_df[pollution_df["Year"] > 2005]
pollution_2006_2010
```

```python
pollution_county_ct = pollution_2006_2010["County"].unique().tolist()
print(pollution_county_ct, len(pollution_county_ct))
```

```python
diseases = mortality_2006_2010["ICD-10 113 Cause List"].unique().tolist()
print(diseases, len(diseases))
```

```python
pollution_2006_2010.to_csv("Resources/pollution_2006_2010.csv")
```

```python
def common_county(x, y):
    x_set = set(x)
    y_set = set(y)
    
    if len(x_set.intersection(y_set)) > 0:
        return x_set.intersection(y_set)
    else:
        return("No common county")
    
same_county = common_county(mortality_county_ct, pollution_county_ct)
print(same_county, len(same_county))
```

```python
# Making sure there are the same number of counties in Pollution and Mortality Data
mortality_2006_2010_clean = mortality_2006_2010[mortality_2006_2010["County"].isin(same_county)]
mortality_2006_2010_clean
```

```python
mortality_2006_2010_clean.to_csv("Resources/mortality_2006_2010_clean.csv")
```
