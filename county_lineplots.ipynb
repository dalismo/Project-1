```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
mortality_ca = pd.read_csv("Resources/mortality_2006_2010_clean.csv")
pollution_ca = pd.read_csv("Resources/pollution_2006_2010.csv")
```

```python
mortality_ca
```

```python
pollution_ca
```

```python
# Influenza
influenza_pneumonia = mortality_ca[mortality_ca["ICD-10 113 Cause List"] == 
                                         "Influenza and pneumonia (J09-J18)"]
influenza_pneumonia.head()
```

```python
grouped_influenza_pneumonia = influenza_pneumonia.groupby(["County", "Year"])["Deaths"].agg([np.sum])
grouped_influenza_pneumonia
```

```python
influenza_pneumonia_alameda = grouped_influenza_pneumonia.filter(like="Alameda", axis=0)
influenza_pneumonia_alameda1 = influenza_pneumonia_alameda.rename(columns={"sum": "Death"})
influenza_pneumonia_alameda1
```

```python
pollutants_alameda = pollution_ca[pollution_ca["County"] == "Alameda"]
pollutants_alameda
```

```python
co_pollutants_alameda = pollutants_alameda.groupby(["County", "Year"])["CO AQI"].agg([np.sum])
co_pollutants_alameda1 = co_pollutants_alameda.rename(columns={"sum": "CO AQI"})
co_pollutants_alameda1
```

```python
merged_alameda_data = pd.merge(co_pollutants_alameda1, influenza_pneumonia_alameda1, on="Year")
```

```python
merged_alameda_data1 = merged_alameda_data.reset_index()
merged_alameda_data1
```

```python
x = merged_alameda_data1["Year"]
y1 = merged_alameda_data1["CO AQI"]
y2 = merged_alameda_data1["Death"]
fig, ax = plt.subplots()
ax.plot(x, y1, color="red", marker="o")

ax2 = ax.twinx()
ax2.plot(x, y2, marker="o")

ax.xaxis.set_ticks(x)
ax.set_xlabel("Year")
ax.set_ylabel("Total CO AQI (ppm)")
ax2.set_ylabel("Number of Deaths")
plt.title("Number of Influenza & Pneumonia Deaths in Alameda, CA\nvs Total CO AQI | 2006-2010\n")

plt.show()
plt.tight_layout()
```
