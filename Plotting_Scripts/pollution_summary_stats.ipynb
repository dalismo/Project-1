```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = "Resources/California_Pollution.csv"
pollution = pd.read_csv(filepath)
```

```python
pollution.head()
```

```python
pollution_clean_data = pollution.drop(["Unnamed: 0", 
                                       "State Code", 
                                       "County Code", 
                                       "Site Num", 
                                       "Address", 
                                       "State"], axis=1)
```

```python
pollution_clean_data.head()
```

```python
pollution_clean_data1 = pollution_clean_data
pollution_clean_data1["Date Local"] = pd.DatetimeIndex(pollution_clean_data["Date Local"]).year
pollution_clean_data1["Date Local"]
```

```python
pollution_clean_data2 = pollution_clean_data1.rename(columns={"Date Local": "Year"})
pollution_clean_data2.head()
```

```python
# Function for line plot over bar plot for mean, max and min of AQI for each county
def line_bar_plot(s):
    ax = s["mean"].plot(x='mean',linestyle='-', 
                        marker='o', color="black", 
                        linewidth=1, legend=True)

    s[["amax", "amin"]].plot(kind="bar", ax=ax, 
                             figsize=(15,5), grid=True)
    
# how to use above function
# line_bar_plot(grouped_pollution_by_no2)
# plt.ylabel("Air Quality Index (AQI) - Parts Per Billion")
# plt.title("NO2 Air Quality Index of 18 California Counties")
# plt.tight_layout()
# plt.show()
```

```python
# Air Quality Index by NO2
grouped_pollution_by_no2 = pollution_clean_data2.groupby(["County", "Year"])["NO2 AQI"].agg([np.mean, 
                                                                                            np.max, 
                                                                                            np.min])
```
